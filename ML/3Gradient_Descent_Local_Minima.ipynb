Gradient Descent to Find Local Minima of y = (x + 3)^2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function y = (x + 3)^2
def f(x):
    return (x + 3) ** 2
```

```python
# Derivative of the function f'(x) = 2 * (x + 3)
def df(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent Function
def gradient_descent(starting_point, learning_rate, epochs):
    x = starting_point
    history = [x]

    for i in range(epochs):
        grad = df(x)                # Calculate gradient
        x = x - learning_rate * grad  # Update rule
        history.append(x)

        if i % 10 == 0:
            print(f'Iteration {i}: x = {x:.4f}, f(x) = {f(x):.4f}')

    return x, history

# Parameters
x_start = 2          # Starting point
alpha = 0.1          # Learning rate
epochs = 50          # Number of iterations

```

```python
# Run gradient descent
x_min, path = gradient_descent(x_start, alpha, epochs)
print('Local Minima found at:')
print(f'x = {x_min:.4f}')
print(f'f(x) = {f(x_min):.4f}')
```

```python
# Visualization
x_values = np.linspace(-6, 4, 100)
y_values = f(x_values)

plt.figure(figsize=(8,5))
plt.plot(x_values, y_values, label='y = (x + 3)^2')
plt.scatter(path, [f(x) for x in path], color='red', label='Descent Path')
plt.scatter(x_min, f(x_min), color='green', s=100, label='Local Minima')
plt.title('Gradient Descent: Finding Local Minima')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
